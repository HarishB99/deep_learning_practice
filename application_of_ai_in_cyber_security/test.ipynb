```python
import os
import matplotlib.pyplot as plt
from keras.preprocessing import image
from PIL import Image
```

```python
def convert(filename, imgname):
    try:
        # target_size is of the format: (img_height, img_width)
        img = image.load_img(filename, color_mode='grayscale', target_size=(1024, 1024))
        img.save(imgname)
    except EnvironmentError as error:
        print('Error in test.convert(): ', error.strerror)
```

```python
image_dir = os.path.abspath('./images')
ramnit_image_dir = os.path.join(image_dir, '1_ramnit')
lollipop_image_dir = os.path.join(image_dir, '2_lollipop')
kelihos_ver3_image_dir = os.path.join(image_dir, '3_kelihos_ver3')
vundo_image_dir = os.path.join(image_dir, '4_vundo')
simda_image_dir = os.path.join(image_dir, '5_simda')
tracur_image_dir = os.path.join(image_dir, '6_tracur')
kelihos_ver1_image_dir = os.path.join(image_dir, '7_kelihos_ver1')
obfuscator_acy_image_dir = os.path.join(image_dir, '8_obfuscator_acy')
gatak_image_dir = os.path.join(image_dir, '9_gatak')

image_modified_dir = os.path.abspath('./images_modified')
if not os.path.exists(image_modified_dir):
    os.mkdir(image_modified_dir)
    
ramnit_image_modified_dir = os.path.join(image_modified_dir, '1_ramnit')
if not os.path.exists(ramnit_image_modified_dir):
    os.mkdir(ramnit_image_modified_dir)
    
lollipop_image_modified_dir = os.path.join(image_modified_dir, '2_lollipop')
if not os.path.exists(lollipop_image_modified_dir):
    os.mkdir(lollipop_image_modified_dir)
    
kelihos_ver3_image_modified_dir = os.path.join(image_modified_dir, '3_kelihos_ver3')
if not os.path.exists(kelihos_ver3_image_modified_dir):
    os.mkdir(kelihos_ver3_image_modified_dir)
    
vundo_image_modified_dir = os.path.join(image_modified_dir, '4_vundo')
if not os.path.exists(vundo_image_modified_dir):
    os.mkdir(vundo_image_modified_dir)
    
simda_image_modified_dir = os.path.join(image_modified_dir, '5_simda')
if not os.path.exists(simda_image_modified_dir):
    os.mkdir(simda_image_modified_dir)
    
tracur_image_modified_dir = os.path.join(image_modified_dir, '6_tracur')
if not os.path.exists(tracur_image_modified_dir):
    os.mkdir(tracur_image_modified_dir)
    
kelihos_ver1_image_modified_dir = os.path.join(image_modified_dir, '7_kelihos_ver1')
if not os.path.exists(kelihos_ver1_image_modified_dir):
    os.mkdir(kelihos_ver1_image_modified_dir)
    
obfuscator_acy_image_modified_dir = os.path.join(image_modified_dir, '8_obfuscator_acy')
if not os.path.exists(obfuscator_acy_image_modified_dir):
    os.mkdir(obfuscator_acy_image_modified_dir)
    
gatak_image_modified_dir = os.path.join(image_modified_dir, '9_gatak')
if not os.path.exists(gatak_image_modified_dir):
    os.mkdir(gatak_image_modified_dir)
    
```
